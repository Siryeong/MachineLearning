```python
### gausian plot
def gausian(data):
  n = data.size
  uml = np.mean(data)
  vml = np.var(data)
  sigma = np.std(data)
  var = n*vml/(n-1) # un bias
  return (1/np.sqrt(2*np.pi*var))*np.exp(-(1/(2*var))*np.power(data-uml, 2))

plt.scatter(h, gausian(h), s=0.1)
plt.show()
plt.scatter(hm, gausian(hm), s=0.1)
plt.show()
plt.scatter(hf, gausian(hf), s=0.1)
plt.show()

plt.scatter(w, gausian(w), s=0.1)
plt.show()
plt.scatter(wm, gausian(wm), s=0.1)
plt.show()
plt.scatter(wf, gausian(wf), s=0.1)
plt.show()
```

```python
### kernel density estimation
def KDE(data, bandwidth):
  n = data.size
  kde = np.array([])
  for x in data:
    px = 1/(n*bandwidth)*data[((x - bandwidth/2) < data) & (data < (x + bandwidth/2))].size
    kde = np.append(kde, px)
  return kde

bandwidth = 10
kde_h = KDE(h, bandwidth)
kde_hm = KDE(hm, bandwidth)
kde_hf = KDE(hf, bandwidth)

kde_w = KDE(w, bandwidth)
kde_wm = KDE(wm, bandwidth)
kde_wf = KDE(wf, bandwidth)
```

```python
plt.scatter(h, kde_h, s=0.5)
plt.show()
plt.scatter(hm, kde_hm, s=0.5)
plt.show()
plt.scatter(hf, kde_hf, s=0.5)
plt.show()

plt.scatter(w, kde_w, s=0.5)
plt.show()
plt.scatter(wm, kde_wm, s=0.5)
plt.show()
plt.scatter(wf, kde_wf, s=0.5)
plt.show()
```
